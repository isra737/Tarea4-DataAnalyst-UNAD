2. Preprocesamiento de datos

```python
# Importar librerías para preprocesamiento
import pandas as pd  # Manejo de datos
from sklearn.preprocessing import StandardScaler  # Estandarización de variables

# Cargar el dataset desde el archivo CSV
Cliente = pd.read_csv('C:/Users/Family/Videos/Mall_Customers.csv')  # Lee el archivo 'Mall_Customers.csv' se debe copiar la ruta completa

# Mostrar las columnas disponibles para verificar la presencia de 'CustomerID'
print("\nColumnas en el dataset:")
print(Cliente.columns.tolist())  # Imprime la lista de columnas

# Eliminar la columna 'CustomerID' si existe
if 'CustomerID' in data.columns:
    data = Cliente.drop('CustomerID', axis=1)  # Elimina la columna 'CustomerID'
    print("\nColumna 'CustomerID' eliminada.")
else:
    print("\nLa columna 'CustomerID' no está presente en el dataset.")

# Codificar la variable categórica 'Gender' (Male=0, Female=1)
Cliente['Gender'] = Cliente['Gender'].map({'Male': 0, 'Female': 1})  # Convierte 'Gender' a valores numéricos

# Verificar nuevamente si hay valores nulos
print("\nValores nulos después del preprocesamiento:")
print(Cliente.isnull().sum())  # Muestra la suma de valores nulos por columna

# Estandarizar las variables numéricas para que tengan media 0 y desviación estándar 1
scaler = StandardScaler()  # Inicializar el escalador
scaled_cliente = scaler.fit_transform(Cliente[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']])  # Estandarizar variables
scaled_df = pd.DataFrame(scaled_cliente, columns=['Age', 'Annual Income (k$)', 'Spending Score (1-100)'])  # Crear dataframe con datos escalados
scaled_df['Gender'] = Cliente['Gender'].values  # Agregar la columna 'Gender' sin escalar

# Mostrar las primeras filas del dataframe preprocesado
print("\nPrimeras filas del dataframe preprocesado:")
print(scaled_df.head())  # Imprime las primeras filas
```

3. Selección de Características

```python
# Seleccionar las características más relevantes para hierarchical clustering
X = scaled_df[['Annual Income (k$)', 'Spending Score (1-100)']]  # Selecciona 'Annual Income' y 'Spending Score'

# Mostrar las primeras filas de las características seleccionadas
print("Características seleccionadas para Hierarchical Clustering:")
print(X.head())  # Imprime las primeras filas del dataframe X
```

4. Entrenamiento del modelo para configurar hiperparámetros

```python
# Importar librerías para hierarchical clustering y visualización
from scipy.cluster.hierarchy import dendrogram, linkage  # Funciones para dendrograma
from sklearn.cluster import AgglomerativeClustering  # Algoritmo de clustering jerárquico
import matplotlib.pyplot as plt  # Creación de gráficos

# Configurar estilo de gráficos
plt.style.use('seaborn-v0_8')  # Estilo 'seaborn-v0_8' para gráficos más atractivos

# Generar el dendrograma para determinar el número de clusters
linked = linkage(X, method='ward')  # Calcular la matriz de enlace usando el método 'ward'
plt.figure(figsize=(10, 7))  # Crear figura de 10x7 pulgadas
dendrogram(linked, truncate_mode='level', p=5)  # Graficar dendrograma con 5 niveles
plt.title('Dendrograma para Hierarchical Clustering')  # Título del gráfico
plt.xlabel('Muestras')  # Etiqueta del axe X
plt.ylabel('Distancia')  # Etiqueta del eje Y
plt.show()  # Mostrar el gráfico

# Entrenar el modelo de clustering jerárquico con 5 clusters
hierarchical = AgglomerativeClustering(n_clusters=5, metric='euclidean', linkage='ward')   # Inicializar modelo
labels_hierarchical = hierarchical.fit_predict(X)  # Entrenar y obtener etiquetas de clusters

# Mostrar las primeras 10 etiquetas de clusters
print("\nPrimeras etiquetas de clusters:")
print(labels_hierarchical[:10])  # Imprime las primeras etiquetas
```

5. Evalación de desempeño con métricas como Coeficiente de Silhouette, Índice de Calinski-Harabasz

```python
# Importar librerías para métricas de evaluación
from sklearn.metrics import silhouette_score, calinski_harabasz_score  # Métricas de clustering

# Calcular el Coeficiente de Silhouette (mide la separación entre clusters)
silhouette_hierarchical = silhouette_score(X, labels_hierarchical)  # Calcular Silhouette

# Calcular el Índice de Calinski-Harabasz (mide la compactación y separación)
calinski_hierarchical = calinski_harabasz_score(X, labels_hierarchical)  # Calcular Calinski-Harabasz

# Mostrar los resultados de las métricas
print("\nEvaluación Hierarchical Clustering:")
print(f"Coeficiente de Silhouette: {silhouette_hierarchical:.3f}")  # Imprimir Silhouette
print(f"Índice de Calinski-Harabasz: {calinski_hierarchical:.3f}")  # Imprimir Calinski-Harabasz
```

6. Graficas para visualizar los datos del modelo 

```python
# Importar librerías para visualización
import pandas as pd  # Manejo de datos
import seaborn as sns  # Visualizaciones avanzadas
import matplotlib.pyplot as plt  # Creación de gráficos

# Configurar estilo de gráficos
plt.style.use('seaborn-v0_8')  # Estilo 'seaborn-v0_8' para gráficos más atractivos

# --- Gráfico 1: Dispersión de clusters ---
# Graficar los clusters generados por hierarchical clustering
plt.figure(figsize=(10, 6))  # Crear figura de 10x6 pulgadas
sns.scatterplot(x=X['Annual Income (k$)'], y=X['Spending Score (1-100)'], hue=labels_hierarchical, palette='deep')  # Gráfico de dispersión coloreado por cluster
plt.title('Clusters Hierarchical Clustering (Dispersión)')  # Título del gráfico
plt.xlabel('Ingreso Anual (k$) (Estandarizado)')  # Etiqueta del eje X
plt.ylabel('Puntaje de Gasto (Estandarizado)')  # Etiqueta del eje Y
plt.show()  # Mostrar el gráfico

# --- Gráfico 2: Barras apiladas de distribución por género ---
# Crear un DataFrame con las etiquetas de clusters y género
cluster_gender_df = pd.DataFrame({'Cluster': labels_hierarchical, 'Gender': data['Gender']})
# Contar la cantidad de hombres (0) y mujeres (1) por cluster
gender_counts = cluster_gender_df.groupby(['Cluster', 'Gender']).size().unstack(fill_value=0)
# Renombrar las columnas para claridad
gender_counts.columns = ['Male', 'Female']

# Graficar barras apiladas
plt.figure(figsize=(10, 6))  # Crear figura de 10x6 pulgadas
gender_counts.plot(kind='bar', stacked=True, color=['#1f77b4', '#ff7f0e'])  # Barras apiladas con colores distintivos
plt.title('Distribución de Género por Cluster (Hierarchical)')  # Título del gráfico
plt.xlabel('Cluster')  # Etiqueta del eje X
plt.ylabel('Número de Clientes')  # Etiqueta del eje Y
plt.legend(title='Género')  # Mostrar leyenda con título
plt.xticks(rotation=0)  # Mantener etiquetas del eje X horizontales
plt.show()  # Mostrar el gráfico

# --- Gráfico 3: Boxplot de características por cluster ---
# Crear un DataFrame con las etiquetas de clusters y datos originales
cluster_data = X.copy()  # Copiar los datos estandarizados
cluster_data['Cluster'] = labels_hierarchical  # Agregar etiquetas de clusters

# Crear subgráficos para boxplots
plt.figure(figsize=(15, 5))  # Crear figura de 15x5 pulgadas

# Boxplot para Annual Income
plt.subplot(1, 2, 1)  # Primer subgráfico
sns.boxplot(x='Cluster', y='Annual Income (k$)', data=cluster_data)  # Boxplot por cluster
plt.title('Ingreso Anual por Cluster (Hierarchical)')  # Título del gráfico
plt.xlabel('Cluster')  # Etiqueta del eje X
plt.ylabel('Ingreso Anual (k$) (Estandarizado)')  # Etiqueta del eje Y

# Boxplot para Spending Score
plt.subplot(1, 2, 2)  # Segundo subgráfico
sns.boxplot(x='Cluster', y='Spending Score (1-100)', data=cluster_data)  # Boxplot por cluster
plt.title('Puntaje de Gasto por Cluster (Hierarchical)')  # Título del gráfico
plt.xlabel('Cluster')  # Etiqueta del eje X
plt.ylabel('Puntaje de Gasto (Estandarizado)')  # Etiqueta del eje Y

plt.tight_layout()  # Ajustar espaciado entre subgráficos
plt.show()  # Mostrar los gráficos
```
